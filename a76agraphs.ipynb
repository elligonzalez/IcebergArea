```python
# Example data with dates as a dictionary
data_dict = numsdata
# Convert the dictionary to a pandas DataFrame
dates = pd.to_datetime(list(data_dict.keys()))
data_values = pd.Series(list(data_dict.values()), index=dates)

# Define the window size for the moving average
window_size = 5

# Compute the moving average
moving_avg = data_values.rolling(window=window_size).mean()

# Define a threshold (range) around the moving average
threshold = 750

# Create a filtered DataFrame, keeping data points within the range, and keeping initial data without enough values for moving average
filtered_data = pd.Series(index=dates)

# Apply filtering for points with enough data for the moving average
for i in range(len(data_values)):
    if i < window_size - 1:
        # Keep the data for the first window_size-1 points since moving average isn't calculated yet
        filtered_data.iloc[i] = data_values.iloc[i]
    else:
        # Filter based on the range around the moving average
        if (data_values.iloc[i] >= (moving_avg.iloc[i] - threshold)) and (data_values.iloc[i] <= (moving_avg.iloc[i] + threshold)):
            filtered_data.iloc[i] = data_values.iloc[i]


# Remove NaN values from filtered data
filtered_data.dropna(inplace=True)

filtered_data_dict = {date.strftime('%Y-%m-%d'): value for date, value in filtered_data.items()}
```
